```python
# Creating a function update the data in the dataframe
def api_runner():
    url = 'https://pro-api.coinmarketcap.com/v1/cryptocurrency/listings/latest'
    parameters = {
    'start':'1',
    'limit':'15',
    'convert':'USD'
    }
    headers = {
    'Accepts': 'application/json',
    'X-CMC_PRO_API_KEY': '2de120d3-aa7f-4d00-800f-75b125168694',
    }

    session = Session()
    session.headers.update(headers)

    try:
        response = session.get(url, params=parameters)
        data = json.loads(response.text)
        #print(data)
    except (ConnectionError, Timeout, TooManyRedirects) as e:
        print(e)
    
    df2 = pd.json_normalize(data['data']) # this normalizes the data and puts it in a dataframe

    df2['timestamp'] = pd.to_datetime('now') # this adds a timestamp to our data to show when its last updated
    df = df2.append(df2)
```

```python
#import os
#from time import time
#from time import sleep
# the above enables us to use the operating system to track the time
for i in range(333): # the api has a limit of 333 calls in a day
    api_runner()
    print('API ruuner succesful')
    sleep(60) #sleep for 1 minute
exit()
```
